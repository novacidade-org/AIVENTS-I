```python
import pandas as pd
```

```python
df = pd.read_json("all_events.json")
df.to_csv()
```

```python
df
```

```python
df.info()
```

```python
# start e end date não têm horas dif de 00:00:00 -> tirar as horas e passar para date type
df["startDate"] = pd.to_datetime(df["startDate"], errors="coerce").dt.normalize()
df["endDate"] = pd.to_datetime(df["endDate"], errors="coerce").dt.normalize()
```

```python
# separar city -> city_id + city_name
city_cols = (
    pd.json_normalize(df["city"])        # -> columns: ['id','title']
      .rename(columns={"id": "city_id", "title": "city_name"})
)
df = df.join(city_cols)
```

```python
df.head()
```

```python
# separar category -> category_id + category_name
category_cols = (
    pd.json_normalize(df["category"])
      .rename(columns={"id": "category_id", "title": "category_title"})
)
df = df.join(category_cols)
```

```python
df["monday_start_date"] = df["startDate"].dt.to_period("W-SUN").dt.start_time
df["monday_end_date"] = df["endDate"].dt.to_period("W-SUN").dt.start_time
```

```python
df["title"] = df["title"].str.strip()
```

```python
# remove unnecessary columns
final_df = df.drop(columns=["id", "image", "city", "category"])
# manter link do evento - chatbot pode partilhar o link com o user para mais informações
```
